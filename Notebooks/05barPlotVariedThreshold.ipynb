# Varying Distance Threshold

Here I'm collecting into a dataframe the number of close/far direct/transfer trips for different thresholds of "close". The data is plotted on the website

```python
results = pd.DataFrame(columns = ['dist','name','x','y'])
threshs = [0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.85,0.9,0.95,1.0]
for dist_thresh in threshs:
    print dist_thresh
    
    total_trips = 0
    total_direct = 0
    total_single = 0
    total_multiple = 0

    total_close = 0
    total_close_direct = 0
    total_close_single = 0
    total_close_multiple = 0
    
    for i in xrange(1,13):
        print i
        trip_data = pd.read_csv('../Data/Analyzed/trip_data_analyzed%d.csv'%i,usecols=['dist_start','dist_end','num_transfers'])

        total_trips += len(trip_data)
        close_groups = trip_data[(trip_data['dist_start']<dist_thresh) & (trip_data['dist_end']<dist_thresh)].groupby('num_transfers').size()
        far_groups = trip_data[(trip_data['dist_start']>=dist_thresh) | (trip_data['dist_end']>=dist_thresh)].groupby('num_transfers').size()

        total_direct += close_groups['direct'] + far_groups['direct']
        total_single += close_groups['single transfer'] + far_groups['single transfer']
        total_multiple += close_groups['multiple transfers'] + far_groups['multiple transfers']

        total_close_direct += close_groups['direct']
        total_close_single += close_groups['single transfer']
        total_close_multiple += close_groups['multiple transfers']
        total_close += close_groups.sum()

    total_far_direct = total_direct - total_close_direct
    total_far_single = total_single - total_close_single
    total_far_multiple = total_multiple - total_close_multiple

    #x=0 -> close, x=1 -> far
    thresh_results = pd.DataFrame([{'name': 'Direct', 'x': 0, 'y': float(total_close_direct)/total_trips, 'dist': dist_thresh},
                 {'name': 'Single Transfer', 'x': 0, 'y': float(total_close_single)/total_trips, 'dist': dist_thresh},
                 {'name': 'Multiple Transfers', 'x': 0, 'y': float(total_close_multiple)/total_trips, 'dist': dist_thresh},
                 {'name': 'Direct', 'x': 1, 'y': float(total_far_direct)/total_trips, 'dist': dist_thresh},
                 {'name': 'Single Transfer', 'x': 1, 'y': float(total_far_single)/total_trips, 'dist': dist_thresh},
                 {'name': 'Multiple Transfers', 'x': 1, 'y': float(total_far_multiple)/total_trips, 'dist': dist_thresh}]
                )
    results = results.append(thresh_results)
    
    assert (total_trips == total_close_direct + total_close_single + total_close_multiple + total_far_direct + total_far_single + total_far_multiple)
    print total_trips
```

```python
results.to_csv('../Data/Analyzed/dist_transfer.csv')
```
