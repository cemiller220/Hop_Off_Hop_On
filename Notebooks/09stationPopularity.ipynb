```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['savefig.dpi'] = 144

import pandas as pd
from ast import literal_eval
import datetime
import time

import warnings
warnings.filterwarnings('ignore')
```

Value assigned to each station for each trip is combination of different things:

- If station is starting or ending station for trip:
    - initial_value = 1

- If station is transfer station:
    - inital_value = 1/(# of convenient transfer stations for trip)
    - "convenient transfer stations" are defined as the stations that require the fewest number of stops
    
- The value is divided by the number of different lines those passengers could take between those two stations
- The value is multiplied by the number of passengers from that taxi trip

```python
stops = pd.read_csv('../Data/stops_updated.txt',usecols=['stop_id','parent_station','lines'])
stops['lines'] = stops['lines'].apply(literal_eval)
stops = stops[stops['parent_station'].isnull()][['stop_id','lines']]
```

```python
def day_of_week(row):
    days = {0: 'Weekday', 1: 'Weekday', 2: 'Weekday', 3: 'Weekday', 4: 'Weekday', 5: 'Weekend', 6: 'Weekend'}
    return days[row['pickup_datetime'].weekday()]
```

```python
def get_time(row, name):
    return row['%s_datetime'%name].time()
```

```python
def lines1(row):
    return len(set(row['lines_start']).intersection(row['lines_start_transfer']))

def lines2(row):
    return len(set(row['lines_end']).intersection(row['lines_end_transfer']))
```

```python
def parse_time(row,name,group):
    hour = row['%s_datetime'%name].hour
    minute_group = row['%s_datetime'%name].minute/group
    return '%d %d'%(hour, minute_group)
```

```python
def to_time(row):
    split = row['hour_minute'].split()
    return datetime.time(int(split[0]),int(split[1])*10,0)
```

```python
for i in xrange(1,13):
    print i

    #read in data
    print 'read in data'
    start = time.time()
    trip_data = pd.read_csv('../Data/Convenience/trip_data_convenience%d.csv'%i,usecols=['passenger_count','pickup_datetime','stop_id_start','lines_start','dropoff_datetime','stop_id_end','lines_end','stop_id_start_transfer','stop_id_end_transfer','num_of_stops_1','num_of_stops_2','num_of_stops','value'],parse_dates=['pickup_datetime','dropoff_datetime'],infer_datetime_format=True)
    trip_data['lines_start'] = trip_data['lines_start'].apply(literal_eval)
    trip_data['lines_end'] = trip_data['lines_end'].apply(literal_eval)
    #end = time.time()
    #print 'time:', end-start
    
    #add column for Weekday, or Weekend
    print 'add day'
    #start = time.time()
    trip_data['day'] = trip_data.apply(day_of_week,axis=1)
    #end = time.time()
    #print 'time:', end-start
    
    #add column for transfer datetime, based on ratio of number of stops
    print 'add transfer time'
    #start = time.time()
    trip_data['transfer_datetime'] = (trip_data['pickup_datetime'] + (trip_data['dropoff_datetime']-trip_data['pickup_datetime'])*(trip_data['num_of_stops_1']/trip_data['num_of_stops']))
    #end = time.time()
    #print 'time:', end-start
    
    #get number of possible lines for each part of trip
    print 'get num lines'
    #start = time.time()
    trip_data = trip_data.merge(stops.rename(index=str, columns={"stop_id": "stop_id_start_transfer", 'lines': 'lines_start_transfer'}), on='stop_id_start_transfer')
    trip_data = trip_data.merge(stops.rename(index=str, columns={"stop_id": "stop_id_end_transfer", 'lines': 'lines_end_transfer'}), on='stop_id_end_transfer')
    trip_data['num_lines1'] = trip_data.apply(lines1,axis=1)
    trip_data['num_lines2'] = trip_data.apply(lines2,axis=1)
    #end = time.time()
    #print 'time:', end-start
    
    #split into 4 DFs
    print 'split DFs'
    #start = time.time()
    start_station_data = trip_data[['passenger_count','stop_id_start','num_lines1','day','pickup_datetime']]
    start_transfer_station_data = trip_data[['passenger_count','stop_id_start_transfer','num_lines1','value','day','transfer_datetime']]
    end_transfer_station_data = trip_data[['passenger_count','stop_id_end_transfer','num_lines2','value','day','transfer_datetime']]
    end_station_data = trip_data[['passenger_count','stop_id_end','num_lines2','day','dropoff_datetime']]
    #end = time.time()
    #print 'time:', end-start
    
    #assign initial value to start and end station
    print 'initial value'
    #start = time.time()
    start_station_data['value'] = 1.0
    end_station_data['value'] = 1.0
    #end = time.time()
    #print 'time:', end-start
    
    #calculate final value
    print 'calc final value'
    #start = time.time()
    start_station_data['adjusted_value_start'] = start_station_data['value']/start_station_data['num_lines1']*start_station_data['passenger_count']
    start_transfer_station_data['adjusted_value_start_transfer'] = start_transfer_station_data['value']/start_transfer_station_data['num_lines1']*start_transfer_station_data['passenger_count']
    end_transfer_station_data['adjusted_value_end_transfer'] = end_transfer_station_data['value']/end_transfer_station_data['num_lines2']*end_transfer_station_data['passenger_count']
    end_station_data['adjusted_value_end'] = end_station_data['value']/end_station_data['num_lines2']*end_station_data['passenger_count']
    #end = time.time()
    #print 'time:', end-start
    
    #get hour and minute
    print 'hour and minute'
    #start = time.time()
    start_station_data['hour_minute'] = start_station_data.apply(lambda x: parse_time(x,'pickup',10),axis=1)
    start_transfer_station_data['hour_minute'] = start_transfer_station_data.apply(lambda x: parse_time(x,'transfer',10),axis=1)
    end_transfer_station_data['hour_minute'] = end_transfer_station_data.apply(lambda x: parse_time(x,'transfer',10),axis=1)
    end_station_data['hour_minute'] = end_station_data.apply(lambda x: parse_time(x,'dropoff',10),axis=1)
    #end = time.time()
    #print 'time:', end - start
    
    
    #combine all months
    print 'combine months'
    #start = time.time()
    if i==1:
        start_data = start_station_data
        start_transfer_data = start_transfer_station_data
        end_transfer_data = end_transfer_station_data
        end_data = end_station_data
    else:
        start_data = pd.concat([start_data,start_station_data])
        start_transfer_data = pd.concat([start_transfer_data,start_transfer_station_data])
        end_transfer_data = pd.concat([end_transfer_data,end_transfer_station_data])
        end_data = pd.concat([end_data,end_station_data])
    end = time.time()
    print 'time:', end-start
```

```python
start = time.time()
start_group = pd.DataFrame(start_data.groupby(['stop_id_start','day','hour_minute'])['adjusted_value_start'].sum(),columns=['adjusted_value_start']).reset_index()
start_transfer_group = pd.DataFrame(start_transfer_data.groupby(['stop_id_start_transfer','day','hour_minute'])['adjusted_value_start_transfer'].sum(),columns=['adjusted_value_start_transfer']).reset_index()
end_transfer_group = pd.DataFrame(end_transfer_data.groupby(['stop_id_end_transfer','day','hour_minute'])['adjusted_value_end_transfer'].sum(),columns=['adjusted_value_end_transfer']).reset_index()
end_group = pd.DataFrame(end_data.groupby(['stop_id_end','day','hour_minute'])['adjusted_value_end'].sum(),columns=['adjusted_value_end']).reset_index()
end = time.time()
print 'time:', end-start
```

```python
groups = start_group.rename(index=str, columns={"stop_id_start": "stop_id"}).merge(
    start_transfer_group.rename(index=str, columns={"stop_id_start_transfer": "stop_id"}), on=['stop_id','day','hour_minute'], how='outer').merge(
    end_transfer_group.rename(index=str, columns={"stop_id_end_transfer": "stop_id"}), on=['stop_id','day','hour_minute'], how='outer').merge(
    end_group.rename(index=str, columns={"stop_id_end": "stop_id"}), on=['stop_id','day','hour_minute'], how='outer')
groups = groups.fillna(0)
groups['adjusted_value'] = groups['adjusted_value_start'] + groups['adjusted_value_start_transfer'] + groups['adjusted_value_end_transfer'] + groups['adjusted_value_end']
```

```python
def normalize_per_day(row):
    if row['day'] == 'Weekday':
        return row['adjusted_value']/5.0
    elif row['day'] == 'Weekend':
        return row['adjusted_value']/2.0

groups['adjusted_value'] = groups.apply(normalize_per_day,axis=1)
```

```python
def sec_to_time(row):
    hour = row['seconds'] // 3600
    minute = (row['seconds'] - hour*3600) / 60
    return datetime.time(int(hour), int(minute), 0)
```
