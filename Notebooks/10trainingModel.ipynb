# Training the Model

To predict the crowdedness of a station at a given time, I tried two different models:
- K Nearest Neighbors Regression
- Random Forest Regression

Each individual station for weekdays and weekends were trained separately and stored in a dictionary of fits.

I used 5-fold cross-validation (CV) to determine the optimal parameter for each model (i.e. number of neighbors or max depth). I only trained the model on stations that contained at least 50 data points, meaning the station had to have crowdedness values for at least 1/3 of the day. This corrected for huge mean squared error values that were seen when training using all the stations. I evaluated the error using mean square error, mean absolute error, and median absolute error.

For K Nearest Neighbors, the CV showed an optimal value of 4, but plotting the data showed this is still highly overfitting. This is most likely due to the noisiness of the data. 

For Random Forest, however, the CV showed an optimal value of 6-7, so I chose 6 to minimize any overfitting due to the noise. 

To see the plotted data comparison, see the Plot Fits notebook.

```python
df = pd.read_csv('../Data/Crowdedness/trip_data_crowdedness_new_periodic.csv')
```

```python
df['stop_day'] = df['stop_id'] + '-' + df['day']
```

```python
size = pd.DataFrame(df.groupby(['stop_day']).size(),columns=['size'])
stops = size[size['size'] > 50].reset_index()['stop_day'].values
```

```python
df_train = df[df['stop_day'].isin(stops)].reset_index(drop=True)
```

# K Nearest Neighbors Regression

```python
neighbors = range(3,18)
mse = pd.DataFrame(columns=['fold','neighbors','station','day','mse','mae','med_ae'])
split_inds = {}

for group in df_train.groupby(['stop_id','day']):
        N = len(group[1])
        folds = math.floor(N/5.)
        rand_inds = np.split(np.random.permutation(group[1].index),[folds,2*folds,3*folds,4*folds])
        split_inds[group[0]] = rand_inds


for neighbor in neighbors:
    print neighbor
    fit_dict = {}
    for i in xrange(5):
        for group in df_train.groupby(['stop_id','day']):
            cv_inds = split_inds[group[0]][i]
            train_inds = [ind for x,inds in enumerate(split_inds[group[0]]) for ind in inds if x != i]
            fit_dict[group[0]] = KNeighborsRegressor(n_neighbors=neighbor).fit(group[1].loc[train_inds][Xcols],group[1].loc[train_inds][ycol])
            cv_df = group[1].loc[cv_inds]
            cv_df['predict'] = fit_dict[group[0]].predict(group[1].loc[cv_inds][Xcols])
            mse = mse.append(pd.DataFrame([[i, neighbor, group[0][0], group[0][1], mean_squared_error(cv_df['adjusted_value'],cv_df['predict']),mean_absolute_error(cv_df['adjusted_value'],cv_df['predict']),median_absolute_error(cv_df['adjusted_value'],cv_df['predict'])]],columns=['fold','neighbors','station','day','mse','mae','med_ad']))
```

```python
mse.groupby('neighbors')[['mse']].mean().plot(kind='bar')
```

```python
mse.groupby('neighbors')[['mae','med_ad']].mean().plot(kind='bar')
```

# Random Forest Regression

```python
depths = range(3,11)
mse_forest = pd.DataFrame(columns=['fold','depth','station','day','mse','mae','med_ae'])
split_inds = {}

for group in df_train.groupby(['stop_id','day']):
        N = len(group[1])
        folds = math.floor(N/5.)
        rand_inds = np.split(np.random.permutation(group[1].index),[folds,2*folds,3*folds,4*folds])
        split_inds[group[0]] = rand_inds


for depth in depths:
    print depth
    fit_dict = {}
    for i in xrange(5):
        for group in df_train.groupby(['stop_id','day']):
            cv_inds = split_inds[group[0]][i]
            train_inds = [ind for x,inds in enumerate(split_inds[group[0]]) for ind in inds if x != i]
            fit_dict[group[0]] = RandomForestRegressor(max_depth=depth).fit(group[1].loc[train_inds][Xcols],group[1].loc[train_inds][ycol])
            cv_df = group[1].loc[cv_inds]
            cv_df['predict'] = fit_dict[group[0]].predict(group[1].loc[cv_inds][Xcols])
            mse_forest = mse_forest.append(pd.DataFrame([[i, depth, group[0][0], group[0][1], mean_squared_error(cv_df['adjusted_value'],cv_df['predict']),mean_absolute_error(cv_df['adjusted_value'],cv_df['predict']),median_absolute_error(cv_df['adjusted_value'],cv_df['predict'])]],columns=['fold','depth','station','day','mse','mae','med_ad']))
```

```python
mse_forest.groupby('depth')[['mse','mae','med_ad']].mean()
```
