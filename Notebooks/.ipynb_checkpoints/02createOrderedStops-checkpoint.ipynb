```python
import pandas as pd
```

```python
stops = pd.read_csv('../Data/stops_updated.txt',usecols=['stop_id','stop_name'])
stop_times = pd.read_csv('../Data/stop_times_updated.txt',usecols=['trip_id','stop_id','stop_sequence','arrival_hour'])
trips = pd.read_csv('../Data/trips.txt',usecols=['trip_id','route_id','service_id'])
```

```python
trips_all = trips[['trip_id','route_id']].merge(stops.merge(stop_times,on='stop_id'),on='trip_id')
```

```python
routes = trips_all.groupby('route_id').size().index
```

```python
routes
```

```python
order1 = pd.DataFrame(columns=['stop_id','stop_sequence','count'])

route = '1'
route_trips = trips_all[trips_all['route_id']==route]
for direction in ['N','S']:
    route_trips_dir = route_trips[route_trips['stop_id'].str.contains(direction)]
    stop_counts = pd.DataFrame(route_trips_dir.groupby(['stop_id','stop_sequence']).size(),columns=['count']).reset_index()
    max_idx = stop_counts.groupby('stop_id')['count'].idxmax()
    order1 = order1.append(stop_counts.loc[max_idx].sort_values('stop_sequence'))

    check = pd.DataFrame(order1.groupby('stop_id').size(),columns=['count']).reset_index(drop=True)
    print route,direction,check[check['count']!=1]
    print order1.groupby('count').size()
    
order1 = order1.reset_index(drop=True)
order1['route_id'] = route
order1
```
