```python
trip_data = pd.read_csv('../Data/Convenience/trip_data_convenience1.csv',usecols=['passenger_count','pickup_datetime','stop_id_start','lines_start','dropoff_datetime','stop_id_end','lines_end','stop_id_start_transfer','stop_id_end_transfer','num_of_stops_1','num_of_stops_2','num_of_stops','value'],parse_dates=['pickup_datetime','dropoff_datetime'],infer_datetime_format=True)
trip_data['lines_start'] = trip_data['lines_start'].apply(literal_eval)
trip_data['lines_end'] = trip_data['lines_end'].apply(literal_eval)
```

```python
def day_of_week(row):
    days = {0: 'Weekday', 1: 'Weekday', 2: 'Weekday', 3: 'Weekday', 4: 'Weekday', 5: 'Saturday', 6: 'Sunday'}
    return days[row['pickup_datetime'].weekday()]

trip_data['day'] = trip_data.apply(day_of_week,axis=1)
```

```python
trip_data['transfer_datetime'] = (trip_data['pickup_datetime'] + (trip_data['dropoff_datetime']-trip_data['pickup_datetime'])*(trip_data['num_of_stops_1']/trip_data['num_of_stops']))
```

```python
def get_time(row, name):
    return row['%s_datetime'%name].time()

trip_data['pickup_time'] = trip_data.apply(lambda x: get_time(x,'pickup'), axis=1)
trip_data['transfer_time'] = trip_data.apply(lambda x: get_time(x,'transfer'), axis=1)
trip_data['dropoff_time'] = trip_data.apply(lambda x: get_time(x,'dropoff'), axis=1)
```

```python
def get_time2(row):
    return row['transfer_datetime'].time()

```

```python
def get_time3(row):
    return row['dropoff_datetime'].time()

```

```python
start_station_data = trip_data[['passenger_count','stop_id_start','lines_start','day','pickup_time']]
start_transfer_station_data = trip_data[['passenger_count','stop_id_start_transfer','lines_start','value','day','transfer_time']]
end_transfer_station_data = trip_data[['passenger_count','stop_id_end_transfer','lines_end','value','day','transfer_time']]
end_station_data = trip_data[['passenger_count','stop_id_end','lines_end','day','dropoff_time']]

start_station_data['value'] = 1.0
end_station_data['value'] = 1.0
```

```python
def adjusted_value_start(row):
    return row['value']/len(row['lines_start'])

def adjusted_value_end(row):
    return row['value']/len(row['lines_end'])

start_station_data['adjusted_value'] = start_station_data.apply(adjusted_value_start,axis=1)
start_transfer_station_data['adjusted_value'] = start_transfer_station_data.apply(adjusted_value_start,axis=1)
end_transfer_station_data['adjusted_value'] = end_transfer_station_data.apply(adjusted_value_end,axis=1)
end_station_data['adjusted_value'] = end_station_data.apply(adjusted_value_end,axis=1)
```
